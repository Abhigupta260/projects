As we seen to doing saveral times z score not able to get data without outliers 

so trying to use IQR method to remove outliers

```python
def removeOutlier(col):
    sorted(col)
    qua1,qua2=col.quantile([0.25,0.75])
    IQR=qua2 - qua1
    lowerRange = qua1- (1.5*IQR)
    upperRange = qua2+(1.5*IQR)
    return lowerRange, upperRange
```

```python
lowScore, highScore= removeOutlier(df1["Standard Error"])
df1["Standard Error"]=np.where(df1["Standard Error"]>highScore,highScore,df1["Standard Error"])
df1["Standard Error"]=np.where(df["Standard Error"]<lowScore,lowScore,df1["Standard Error"])
```
