# Removing outliers & Skew 

```python
ds_new.plot(kind="box",subplots=True,layout=(2,4))

#there is oulier in almost all columns
```

```python
ds_new.skew()
#have to correct skewness as well (values are more than +0.5 and less than -0.5)
```

```python
#Removing the Outliers through IQR method
def removeOutlier(col):
    sorted(col)
    qua1,qua2=col.quantile([0.25,0.75])
    IQR=qua2 - qua1
    lowerRange = qua1- (1.5*IQR)
    upperRange = qua2+(1.5*IQR)
    return lowerRange, upperRange
```

```python
columns=["Diameter","Height","Whole weight","Shucked weight", "Viscera weight", "Shell weight","Rings"]
for i in ds_new[columns]:
    lowScore, highScore= removeOutlier(ds_new[i])
    ds_new[i]=np.where(ds_new[i]>highScore,highScore,ds_new[i])
    ds_new[i]=np.where(ds_new[i]<lowScore,lowScore,ds_new[i])
```

```python
ds_new.plot(kind="box",subplots=True,layout=(2,4))
#outlier is removed now
```

```python
ds_new.skew()

#Skewness almost resolved
```

# Train and Test

```python
# creating in for of input (x) and output (y)

x=ds.drop(columns=["Rings"])
y=ds[["Rings"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=45)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python

lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
#checking the score of model 
lm.score(x_train,y_train)
# model is 56.75% accurate
```

```python
pred=lm.predict(x_test)
print("Predicted value and actual value", pred,y_test)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
#finding r2 score and mean squar error in model 

print("erroe:")

print("Mean absolute error:", mean_absolute_error(y_test, pred))
print("Mean squared error:", mean_squared_error(y_test,pred))

print("Root Mean Squared Error:",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
print(r2_score(y_test,pred))
```
